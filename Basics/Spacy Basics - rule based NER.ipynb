```python
import json
```

```python
with open("hpchapter1.txt","r",encoding="utf-8") as f:
    text = f.read().split("\n\n")
    print(text[0])
```

```python
characters_name = []
with open("hp_characters.json","r",encoding="utf-8") as f:
    characters = json.load(f)
    print(characters[0])
    
for character in characters:
    names = character.split()
    for name in names:
        if "and" != name and "the" != name and 'The' != name :
            name = name.replace(",","").strip()
            characters_name = characters_name + [name]
            
print(characters_name)
```
